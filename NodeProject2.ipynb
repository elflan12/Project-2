```python
import pandas as pd
import plotly.express as px
from scipy.stats import sem
import tabulate
pd.options.display.max_rows = 15
```

```python
df_2015 = pd.read_csv("https://raw.githubusercontent.com/elflan12/Project-2/main/2015.csv")
df_2016 = pd.read_csv("https://raw.githubusercontent.com/elflan12/Project-2/main/2016.csv")
df_2017 = pd.read_csv("https://raw.githubusercontent.com/elflan12/Project-2/main/2017.csv")
```

```python
# This data, from kaggle, comes from a happiness survey done by the Gallup World Poll 
# in 155 different countries. The happiness scores come from the answers from life
# evaluation questions posed to respondants. The testing covered six subjects; 
# economic production, social support, life expectancy, freedom, absense of 
# corruption, and generosity.  The results for each country are compared to a 
# Dystopia; a hypothetical country made up of the most unhappy people. 
# Dystopia's scores for each subject are the world's lowest national
# average. The Dystopia Residual is how far off the overall dystopian happiness 
# score was from each country's score on a 1-10 scale.
```

```python
#Data Cleaning
```

```python
#2016 and 2017 have confidence intervals instead of standard error, so it's important
#to fix that before combining the datasets. 
```

```python
df_2016['Standard Error'] = df_2016["Upper Confidence Interval"] - df_2016["Lower Confidence Interval"]
df_2016['Standard Error'] = df_2016['Standard Error'].div(3.92)
df_2016.head(5)[df_2016.columns[0:14]]
#assuming a 95% confidence interval
```

```python
df_2016 = df_2016.drop(df_2016.columns[[4]], axis=1)
df_2016.head(5)[df_2016.columns[0:13]]
```

```python
df_2016 = df_2016.drop(df_2016.columns[[1]], axis=1)
```

```python
df_2016 = df_2016.drop(df_2016.columns[[3]], axis=1)
df_2016.head(5)[df_2016.columns[0:13]]
```

```python
df_2017['Standard Error'] = df_2017["Whisker.high"] - df_2017["Whisker.low"]
df_2017['Standard Error'] = df_2017['Standard Error'].div(3.92)
#same for 2017
```

```python
df_2017.head(5)[df_2017.columns[0:13]]
```

```python
df_2017 = df_2017.drop(df_2017.columns[[3, 4]], axis=1)
df_2017.head(5)[df_2017.columns[0:13]]
```

```python
df_2015 = df_2015.drop(df_2015.columns[[1]], axis=1)
df_2015.head(5)[df_2015.columns[0:13]]
#Removing region for all 3 years because 2017 doesn't have it 
```

```python
df_2017.head(5)[df_2017.columns[0:13]]
```

```python
df_2017 = df_2017.drop(df_2017.columns[[3]], axis=1)
```

```python
df_2017 = df_2017.drop(df_2017.columns[[1]], axis=1)
```

```python
#Making the column names uniform for a clean merge
```

```python
print(df_2016.columns)
```

```python
print(df_2017.columns)
```

```python
print(df_2015.columns)
```

```python
df_2015 = df_2015.reindex(columns=['Country', 'Happiness Rank', 'Happiness Score',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual', 'Standard Error'])
```

```python
print(df_2015.columns)
```

```python
print(df_2017.columns)
```

```python
df_2017 = df_2017.reindex(columns=['Country', 'Happiness.Rank', 'Happiness.Score',
       'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.',
       'Freedom', 'Trust..Government.Corruption.', 'Generosity',
       'Dystopia.Residual', 'Standard Error'])
```

```python
print(df_2017.columns)
```

```python
df_2017.head(5)[df_2017.columns[0:11]]
```

```python
df_2017 = df_2017.rename(columns = {'Country':'Country', 'Happiness.Rank':'Happiness Rank', 'Happiness.Score':'Happiness Score',
       'Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 'Family':'Family', 'Health..Life.Expectancy.':'Health (Life Expectancy)',
       'Freedom':'Freedom', 'Trust..Government.Corruption.':'Trust (Government Corruption)', 'Generosity':'Generosity',
       'Dystopia.Residual':'Dystopia Residual', 'Standard Error':'Standard Error'})
```

```python
#Adding a column for year
```

```python
df_2015['Year'] = 2015
df_2015.head(5)[df_2015.columns[0:12]]
```

```python
df_2016['Year'] = 2016
df_2016.head(5)[df_2016.columns[0:12]]
```

```python
df_2016.shape, df_2015.shape
```

```python
df_2017['Year'] = 2017
df_2017.head(5)[df_2017.columns[0:12]]
```

```python
#Merging the dataframes
```

```python
big_df = df_2015.append(df_2016, ignore_index=True, sort=False)
big_df.head(5)[big_df.columns[0:12]]
big_df.shape
```

```python
happy_df = big_df.append(df_2017, ignore_index=True, sort=False)
happy_df.shape
```

```python
happy_df.head(5)[happy_df.columns[0:13]]
```

```python
happy_df_order = happy_df.sort_values(by = 'Happiness Score', ascending = False)
happy_df_order.head(5)[happy_df_order.columns[0:12]]
```
